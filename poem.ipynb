```python
import pandas as pd
traindf = pd.read_csv("poem_train.csv", encoding="utf-8")
traindf
```

```python
testdf = pd.read_csv("poem_test.csv", encoding="utf-8")
testdf
```

```python
# 先try一個
import jieba
s = "青天有月來幾時"
# cut -> [詞1, 詞2]
# 一定要做的事 -> 切 -> 變得像是西方語言
# 我吃蘋果(東方)  我 吃 蘋果(西方)
" ".join(jieba.cut(s))
```

```python
def poem(s):
    s = s.replace("\r", "").replace("\n", "")
    return " ".join(jieba.cut(s))
traindf["內容"] = traindf["內容"].apply(poem)
traindf
```

```python
testdf["內容"] = testdf["內容"].apply(poem)
testdf
```

```python
# scikit-learn: 答案一定要變成整數 0 1 2
trans = {"李白":0, "杜甫":1, "白居易":2}
traindf["作者"] = traindf["作者"].replace(trans)
traindf
```

```python
testdf["作者"] = testdf["作者"].replace(trans)
testdf
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer()
trainvec = vec.fit_transform(traindf["內容"])
# 每一首詩對於每個欄位(詞)的出現次數
print(trainvec)
```

```python
# 每一個欄位到底是哪一個詞
print(vec.vocabulary_)
```

```python
testvec = vec.transform(testdf["內容"])
# 每一首詩對於每個欄位(詞)的出現次數
print(testvec)
```

```python
# 做公式出來
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(trainvec, traindf["作者"])
```

```python
# 驗證一下模型好壞
from sklearn.metrics import accuracy_score
pre = clf.predict(testvec)
print("預測的結果:", list(pre))
print("正確的結果:", list(testdf["作者"]))
print("正確率:", accuracy_score(pre, testdf["作者"]))
```

```python
# 預測APP
rtrans = {0:"李白", 1:"杜甫", 2:"白居易"}
s = input("來首詩吧!")
# 當你要predict的時候, 一定要是一個群集(list, Series)
inputs = [poem(s)]
x = vec.transform(inputs)
pre = clf.predict(x)[0]
print("誰寫的?", rtrans[pre])
```
